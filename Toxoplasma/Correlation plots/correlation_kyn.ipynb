```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a pandas DataFrame
file_path = 'median_distance_kyn_15.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Assuming your CSV has columns named 'variable1' and 'variable2'
variable1 = 'median'  # Replace with the actual column name
variable2 = 'kynurenine'  # Replace with the actual column name

# Calculate correlation between the two variables
correlation_coef = df[variable1].corr(df[variable2])

# Create a scatter plot with regression line
sns.set(style='whitegrid')
plt.figure(figsize=(8, 6))
sns.regplot(x=variable1, y=variable2, data=df, color='green') 
plt.title(f'Correlation Plot between {variable1} and {variable2}\nCorrelation Coefficient: {correlation_coef:.2f}')
plt.xlabel(variable1)
plt.ylabel(variable2)
plt.tight_layout()

# Display the plot
plt.show()
```
