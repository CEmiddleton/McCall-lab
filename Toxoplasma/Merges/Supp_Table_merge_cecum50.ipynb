```r
# Read the three CSV files
annotations <- read.csv("C:/Users/caitl/OneDrive/Desktop/Toxoplasma/supplemental table merges/cecum50/annotations_cecum_50.csv")
spearmankyn <- read.csv("C:/Users/caitl/OneDrive/Desktop/Toxoplasma/supplemental table merges/cecum50/RF_output_correlated_to_kynurenine_cecum_50_updated.csv")

# Merge the first two CSV files, keeping all data (full outer join)
merged_data_12 <- merge(annotations, spearmankyn, by = "features", all = FALSE)


# Save the merged data to a new CSV file
write.csv(merged_data_12, "cecum_50_merge.csv", row.names = FALSE)

# View the first few rows of the merged data
head(merged_data_12)

```
